```python
import pandas as pd
```

```python
bios=pd.read_csv('./bios.csv')
bios.head()
```

```python
cleaned_bios=bios.copy()
```

```python
# to remove • from names of athletes
cleaned_bios['name']=bios['Used name'].str.replace("•"," ")
```

```python

# to seperate "Measurements" column into "Height" and "Weight" columns

cleaned_bios[['Height', 'Weight']] = bios['Measurements'].str.split('/', expand=True)
cleaned_bios['height_in_cm']=pd.to_numeric(cleaned_bios['Height'].str.strip(' cm'),errors='coerce')
cleaned_bios['weight_in_kg']=pd.to_numeric(cleaned_bios['Weight'].str.strip(' kg'),errors='coerce')
```

```python

# to extract date of birth and death 

date_pattern=r'(\d+ \w+ \d{4}|\d{4})';
cleaned_bios['born_date']=cleaned_bios['Born'].str.extract(date_pattern)
cleaned_bios['born_year']=cleaned_bios['Born'].str.extract(r'(\d{4})')

cleaned_bios['born_date']=pd.to_datetime(cleaned_bios['born_date'],format='mixed',errors='coerce')
cleaned_bios['born_year']=pd.to_numeric(cleaned_bios['born_year'])

cleaned_bios['died_date']=cleaned_bios['Died'].str.extract(date_pattern)
cleaned_bios['died_date']=pd.to_datetime(cleaned_bios['died_date'],format='mixed',errors='coerce')



```

```python
# to extract born region, city and country from "Born" column
location_pattern = r'in ([\w\s-]+), ([\w\s-]+) \((\w+)\)'
cleaned_bios[['born_region','born_city','born_country']]=cleaned_bios['Born'].str.extract(location_pattern,expand=True)
```

```python
# to remove unnecessary columns
needed_columns=['athlete_id','name','born_date','born_city','born_region','born_country','NOC','height_in_cm','weight_in_kg','died_date']
cleaned_data=cleaned_bios[needed_columns]
```
